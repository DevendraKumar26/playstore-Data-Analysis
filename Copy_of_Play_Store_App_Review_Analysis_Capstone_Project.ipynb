# **There we had a null values, I am going to leave it as it is. And a 19 for rating is not possible, so i assume it’s a ‘1.9’. So let’s change it and see the distribution value on rating column.**

```python
playstore['Rating'].replace(to_replace=[19.0], value=[1.9],inplace=True)
playstore.Rating.unique()
most_Rat=playstore.Rating.unique()
most_Rat
```

```python
#Let's check where the most of time Rating is given
sns.distplot(playstore.Rating)
```

# ***Here we can seen that most of time Rating is given is between 4 and 5. ***

```python
playstore.head()
```

```python
graph = sns.FacetGrid(playstore, col='Category', palette="Set2",  col_wrap=5, height=4)
graph = (graph.map(sns.distplot, "Rating", hist=True, rug=False, color="g"))
```

# **horizontal is the rating value, and vertically is quantity of the rating.**

```python
# Mean Rating
plt.figure(figsize=(10,12))
mean_rat = playstore.groupby(['Category'])['Rating'].mean().sort_values(ascending=False)
sns.barplot(x=mean_rat, y=mean_rat.index, data=playstore)
plt.show()
```

```python
playstore.head()
```

```python
# Reviews

playstore.Reviews.unique()
```

```python
# inside review there is a value with 3.0M with M stand for million, lets change it so it can be measure as float

Reviews = []

for x in playstore.Reviews:
    x = x.replace('M','00')
    Reviews.append(x)

Reviews = list(map(float, Reviews))
playstore['reviews'] = Reviews
sns.distplot(Reviews)
```

```python
playstore.head()
```

# **This graph is the distribution of total reviews on each app.**



```python
graph = sns.FacetGrid(playstore, col='Category', palette="Set1",  col_wrap=5, height=4)
graph = (graph.map(plt.hist, "Reviews", color="b"))
```

```python
# Total reviews
plt.figure(figsize=(12,12))
sum_rew = playstore.groupby(['Category'])['reviews'].sum().sort_values(ascending=False)
sns.barplot(x=sum_rew, y=sum_rew.index, data=playstore)
```
